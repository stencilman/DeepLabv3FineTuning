```python
from PIL import Image
import glob
from pathlib import Path
```

```python
image_names = Path('NutDataset/Images/').glob('*')
maxh = 0
maxw = 0
for img_name in list(image_names):
    im = Image.open(img_name)
    print(img_name)
    h, w = im.size
    print(im.size)
    if h > maxh:
        maxh = h
    if w > maxw:
        maxw = w
print (maxh)
print (maxw)
    
```

```python
image_names = Path('NutDataset/Images/').glob('*')
mask_names = Path('NutDataset/Masks/').glob('*')

for img_name in list(zip(list(image_names), list(mask_names))):
    im = Image.open(img_name[0])
    mask = Image.open(img_name[1])
    
    dstI = Image.new(mode = "RGB", size = (maxh, maxw))
    dstI.paste(im, (0,0))
    dstI.save('SS'+str(img_name[0]))
    print (dstI.size)
    
    dstM = Image.new(mode = "RGB", size = (maxh, maxw))
    dstM.paste(mask, (0,0))
    dstM.save('SS'+str(img_name[1]))
    print (dstM.size)
    
```
